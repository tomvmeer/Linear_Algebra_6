```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_excel("trump_tweets.xlsx").sample(2500)
data = list(data['Tweet'])
```

```python
# Getting all the actual words out
words = []
for tweet in data:
    for word in tweet.split(' '):
        word = word.replace('"', '').replace('.', '').lower().replace(",", '').replace('(', '').replace(')', '')
        if '//' not in word and word not in words and '@' not in word and "#" not in word:
             words.append(word)
```

```python
# Counting the words
tweetterm = pd.DataFrame(index=words).transpose()
for tweet in data:
    tweet_count = {}
    for word in tweet.split(' '):
        word = word.replace('"', '').replace('.', '').lower().replace(",", '').replace('(', '').replace(')', '')
        if '//' not in word and '@' not in word and "#" not in word:
            if word not in tweet_count.keys():
                tweet_count[word] = 0
            tweet_count[word] += 1
    tweetterm = tweetterm.append(tweet_count, ignore_index=True) 
```

```python
tweetterm = tweetterm.fillna(0)
top = tweetterm[list(tweetterm.sum().sort_values(ascending=False).index)[0:25]]

# A keywordâ€“tweet matrix using 2500 tweets from Donald Trump. The top 25 most used words are used.
top
```

```python
# Words and their occurence
top.sum()
```

```python
words = list(top.columns)
top.columns = [i for i in range(len(words))]
top_matrix = np.matrix(top)
u, s, vh = np.linalg.svd(top_matrix)
```

```python
u1 = u[:,0]*-1
u2 = u[:,1]
v1 = vh.transpose()[:,0]*-1
v2 = vh.transpose()[:,1]
```

```python
u12 = np.concatenate((u1,u2),axis=1).T
c = {}
for i in range(2500):
    x = np.zeros([2500,2500])
    x[i,i] = 1.0
    cj = u12*x
    c[i] = [cj[:,i].item(0), cj[:,i].item(1)]
```

```python
v12 = np.concatenate((v1,v2),axis=1).T
d = {}
for i in range(len(words)):
    x = np.zeros([len(words), len(words)])
    x[i,i] = 1.0
    dj = v12*x
    d[words[i]] = [dj[:,i].item(0), dj[:,i].item(1)]
```

```python
x,y,n = [],[],[]
fig, ax = plt.subplots(figsize=[20,20])
for i in d:
    x.append(d[i][0])
    y.append(d[i][1])
    n.append(i)
ax.scatter(x,y)
for i in range(len(n)):
    ax.annotate(n[i], (x[i], y[i]))
ax.set_xlim([0,0.8])
ax.set_ylim([-1,0.75])
# Plot of words based on svd
```

```python
x,y,n = [],[],[]
fig, ax = plt.subplots(figsize=[20,20])
for i in c:
    x.append(c[i][0])
    y.append(c[i][1])
    n.append(i)
ax.scatter(x,y)
for i in range(len(n)):
    ax.annotate(n[i], (x[i], y[i]))
ax.set_xlim([0,0.125])
ax.set_ylim([-0.2,0.1])
# Plot of tweets based on svd
```
