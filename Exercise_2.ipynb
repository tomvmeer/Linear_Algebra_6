# Define $\hat{G}$
This is a column stochastic matrix. Just run code that uses the G-matrix.

```python
G_hat = G.copy()
for col in range(len(G)):
    for row in range(len(G)):
        print(row, col)
        G_hat[row,col] = G[row,col]/sum(G[:,col])
```

# Now define the Google PageRank A-matrix (Random walk)
The percentage can be changed such that it uses random walk.

```python
p = 0.85
n = 18
A = p * G_hat + (1-p) * ones([n,n]) * 1/n
print(A)

# Check if it is a column-stochastic matrix
print('\nCheck if column-stochastic')
for col in range(len(A)):
    print(sum(A[:,col]))
```

```python
eigenvalues, eigenvectors = eig(A)
eigenvalues
```

```python
for val in range(len(eigenvalues)):
    if eigenvalues[val] == 1.:
        print('yeah')
```
