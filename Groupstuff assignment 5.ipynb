# B) Create 100x100 normally distributed matrix A. compute all singular values (S=svd(A)) and plot on logarithmic scale (semilogy(S)). How many values are >5% of that of the largest?

```python
A = np.random.normal(loc =0, scale =1, size=(100, 100))
#print(A)
U, s, VT = svd(A)
print(s)
plt.semilogy(s)
plt.show()
```

```python
cutOff = 0.05*max(s)
count = 0
for i in s:
    if i > cutOff:
        count += 1
    else:
        pass

print("The approximate rank of A is:", count)
```
